```python
import requests as req
```

```python
url = 'https://es.wikipedia.org/wiki/Pandemia_de_COVID-19_en_Uruguay'
```

```python
r = req.get(url)
```

```python
#!pip install bs4
```

```python
print (r.content[:200])
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.content,'html5lib')
```

```python
print(soup.prettify)
```

```python
atributo_tablas_dic = {'class':'wikitable'}
tablas = soup.find_all("table",atributo_tablas_dic)
```
